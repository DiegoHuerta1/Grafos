Util para hacer comparaciones

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Modelos de Clasificacion

```python
# ver despues de n epochs
n = 0

train_acc_pyG = np.load('.\historiales\pytorchGeom_class_acc_train.npy')[n:]
test_acc_pyG = np.load('.\historiales\pytorchGeom_class_acc_test.npy')[n:]
train_acc_py = np.load('.\historiales\pytorch_class_acc_train.npy')[n:]
test_acc_py = np.load('.\historiales\pytorch_class_acc_test.npy')[n:]

train_loss_pyG = np.load('.\historiales\pytorchGeom_class_loss_train.npy')[n:]
test_loss_pyG = np.load('.\historiales\pytorchGeom_class_loss_test.npy')[n:]
train_loss_py = np.load('.\historiales\pytorch_class_loss_train.npy')[n:]
test_loss_py = np.load('.\historiales\pytorch_class_loss_test.npy')[n:]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 4))


ax1.plot(train_loss_pyG, label = 'train pyG')
ax1.plot(train_loss_py, label = 'train py')
ax1.plot(test_loss_pyG, label = 'test pyG')
ax1.plot(test_loss_py, label = 'test py')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.set_title('Perdida en entrenamiento para clasificacion')


ax2.plot(train_acc_pyG, label = 'train pyG')
ax2.plot(train_acc_py, label = 'train py')
ax2.plot(test_acc_pyG, label = 'test pyG')
ax2.plot(test_acc_py, label = 'test py')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.set_title('Presicion en entrenamiento para clasificacion')

plt.show()
```
